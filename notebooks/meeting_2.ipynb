```python
hash_sample = pd.read_csv('/home/maxim/bi2021spring/small_screen/Hash_Sample_small_screen.txt', 
                          header=None, 
                          sep='\t',
                          names=['general', 'barcodes', 'f0'],
                          usecols=[0, 1, 2])
hash_table_out = pd.read_csv('/home/maxim/bi2021spring/small_screen/Hash_Table_out_small_screen.txt', 
                             header=None, 
                             sep='\t', 
                             names=['f1', 'cell_annotation', 'general', 'f2', 'f3'])
```

## Merge

```python
merged = pd.merge(left=hash_table_out, right=hash_sample, on='general')
merged
```

```python
sep_col = pd.DataFrame(merged['general'].str.split('_').tolist(), 
                       columns = ['treatment', 'dose', 'cell'])
sep_col
```

```python
full_data = pd.concat((sep_col, merged[['cell_annotation', 'barcodes']]), axis=1)
full_data
```

```python
final = full_data[['cell_annotation', 'cell', 'treatment', 'dose', 'barcodes']]
final
```
